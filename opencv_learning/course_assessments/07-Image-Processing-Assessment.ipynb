```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def display_img(img,cmap=None):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

**TASK: Open and display the giaraffes.jpg image that is located in the DATA folder.**

```python
giraffes = cv2.imread("../../resources/DATA/giraffes.jpg")
show_giraffes = cv2.cvtColor(giraffes, cv2.COLOR_BGR2RGB)
giraffes.shape
```

```python
display_img(show_giraffes)
```

**TASK:Apply a binary threshold onto the image.**

```python
rect, binary_thresh = cv2.threshold(show_giraffes, 127, 255, cv2.THRESH_BINARY)
rect
```

```python
display_img(binary_thresh)
```

**TASK: Open the giaraffes.jpg file from the DATA folder and convert its colorspace to  HSV and display the image.**

```python
hsv_giraffes = cv2.imread("../../resources/DATA/giraffes.jpg")
hsv_giraffes = cv2.cvtColor(hsv_giraffes, cv2.COLOR_BGR2HSV)
```

```python
display_img(hsv_giraffes)
```

**TASK: Create a low pass filter with a 4 by 4 Kernel filled with values of 1/10 (0.01) and then use 2-D Convolution to blur the giraffer image (displayed in normal RGB)**

```python
kernel = np.ones(shape=(4, 4), dtype=np.float32) / 10
kernel
```

```python
dst = cv2.filter2D(show_giraffes, -1, kernel)

plt.subplot(1, 2, 1)
plt.imshow(show_giraffes)
plt.title("Original")

plt.subplot(1, 2, 2)
plt.imshow(dst)
plt.title("Blur")
plt.show()
```

**TASK: Create a Horizontal Sobel Filter (sobelx from our lecture) with a kernel size of 5 to the grayscale version of the giaraffes image and then display the resulting gradient filtered version of the image.**

```python
show_giraffes_gray = cv2.cvtColor(show_giraffes, cv2.COLOR_RGB2GRAY)
giraffes_sobelX = cv2.Sobel(show_giraffes_gray, cv2.CV_64F, 1, 0, ksize=5)
show_giraffes_gray.shape, giraffes_sobelX.shape
```

```python
display_img(giraffes_sobelX, cmap='gray')
```

**TASK: Plot the color histograms for the RED, BLUE, and GREEN channel of the giaraffe image. Pay careful attention to the ordering of the channels.**

```python
# BGR
plt.figure(figsize=(10, 6))
for idx, channel in enumerate(['b', 'g', 'r']):
    giraffes_histogram = cv2.calcHist([giraffes], channels=[idx], mask=None, histSize=[256], ranges=[0, 256])
    plt.plot(giraffes_histogram, color=channel, label=channel)
plt.grid()
plt.show()
```
