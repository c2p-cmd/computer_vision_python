**TASK: Use PIL and matplotlib to read and display the ../DATA/00-puppy.jpg image.**

```python
from PIL import Image
import matplotlib.pyplot as plt

doggy_img = Image.open('../../resources/DATA/00-puppy.jpg')

plt.imshow(doggy_img)
plt.axis('off')
plt.title("Puppy")
plt.show()
```

**TASK: Convert the image to a NumPy Array**

```python
doggy_np = np.array(doggy_img)
doggy_np.shape
```

**FINAL TASK: Use slicing to set the RED and GREEN channels of the picture to 0, then use imshow() to show the isolated blue channel**

```python
doggy_np_new = doggy_np.copy()
doggy_np_new.shape
```

```python
doggy_np_new[:, :, 0] = 0
doggy_np_new[:, :, 1] = 0

plt.imshow(doggy_np_new)
plt.axis('off')
plt.title("Puppy - Blue Isolated")
plt.show()
```
