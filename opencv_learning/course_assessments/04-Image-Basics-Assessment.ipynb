#### TASK: Open the *dog_backpack.jpg* image from the DATA folder and display it in the notebook. Make sure to correct for the RGB order.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def read_image(path: str):
    _img = cv2.imread(path)
    _img = cv2.cvtColor(_img, cv2.COLOR_BGR2RGB)
    return _img

doggy_backpack = read_image('../../resources/DATA/dog_backpack.jpg')
doggy_backpack.shape
```

```python
plt.title("Image")
plt.imshow(doggy_backpack)
plt.show()
```

#### TASK: Flip the image upside down and display it in the notebook.

```python
doggy_flipped = doggy_backpack.copy()
doggy_flipped = cv2.flip(doggy_flipped, 0)
plt.title("Upside Down")
plt.imshow(doggy_flipped)
plt.show()
```

#### TASK: Draw an empty RED rectangle around the dogs face and display the image in the notebook.

```python
doggy_rect = doggy_backpack.copy()
doggy_rect = cv2.rectangle(
    img=doggy_rect,
    pt1=(200, 400),
    pt2=(600, 700),
    color=(255, 0, 0),
    thickness=3,
)
plt.title("Face Rect")
plt.imshow(doggy_rect)
plt.show()
```

#### TASK: Draw a BLUE TRIANGLE in the middle of the image. The size and angle is up to you, but it should be a triangle (three sides) in any orientation.

```python
doggy_tri = doggy_backpack.copy()
vertices = np.array([
    [150, 700],
    [400, 350],
    [750, 700],
], dtype=np.int32)
points = np.expand_dims(vertices, 1)
doggy_tri = cv2.polylines(
    doggy_tri,
    [points],
    isClosed=True,
    color=(0, 0, 255),
    thickness=3,
)
plt.title("Face Tri")
plt.imshow(doggy_tri)
plt.show()
```

### BONUS TASK. Can you figure our how to fill in this triangle? It requires a different function that we didn't show in the lecture! See if you can use google search to find it.

[CLICK ME FOR A DIRECT LINK TO THE HINT](https://docs.opencv.org/3.0-beta/modules/imgproc/doc/drawing_functions.html#fillpoly)

```python
doggy_tri_fill = doggy_backpack.copy()
vertices = np.array([
    [150, 700],
    [400, 350],
    [750, 700],
], dtype=np.int32)
points = np.expand_dims(vertices, 1)
doggy_tri = cv2.fillPoly(
    doggy_tri_fill,
    [points],
    color=(0, 0, 255),
)
plt.title("Face Tri")
plt.imshow(doggy_tri)
plt.show()
```
