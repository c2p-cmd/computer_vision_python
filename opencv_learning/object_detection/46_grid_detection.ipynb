```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flat_chess = cv2.imread("../../resources/DATA/flat_chessboard.png")
plt.imshow(flat_chess)
plt.title("Flat Chessboard")
plt.show()
```

```python
found, corners = cv2.findChessboardCorners(flat_chess, (7, 7))
found
```

```python
corners.shape
```

```python
flat_chess_copy = flat_chess.copy()
flat_chess_copy = cv2.drawChessboardCorners(flat_chess_copy, (7, 7), corners, found)
plt.imshow(flat_chess_copy)
plt.title("With Grid Detection")
plt.show()
```

```python
dots = cv2.imread('../../resources/DATA/dot_grid.png')
plt.imshow(dots)
plt.show()
```

```python
found, corners = cv2.findCirclesGrid(dots, (10, 10), cv2.CALIB_CB_SYMMETRIC_GRID)
found
```

```python
corners.shape
```

```python
dots_copy = dots.copy()
dots_copy = cv2.drawChessboardCorners(dots_copy, (10, 10), corners, found)
plt.imshow(dots_copy)
plt.title("With Grid Detection")
plt.show()
```
