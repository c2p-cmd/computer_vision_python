```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("../../resources/DATA/sammy_face.jpg")
plt.imshow(img)
plt.title("Sammy Face")
plt.show()
```

```python
edges = cv2.Canny(image=img, threshold1=127, threshold2=127)
plt.imshow(edges)
plt.title("Edges from Canny")
plt.show()
```

```python
edges = cv2.Canny(image=img, threshold1=127, threshold2=255)
plt.imshow(edges)
plt.title("Edges from Canny")
plt.show()
```

```python
median_val = np.median(img)
median_val
```

```python
# Lower Threshold is 0 or 70% of the median value which is greater
# Upper threshold is 130% of median or max value of 255, which is smaller
lower = int(max(0, 0.7 * median_val))
upper = int(min(255, 1.3*median_val))

print(f"Lower: {lower}, Upper: {upper}")

edges = cv2.Canny(image=img, threshold1=lower, threshold2=upper)
plt.imshow(edges)
plt.title("Edges with formula")
plt.show()
```

```python
blurred_img = cv2.blur(img, ksize=(5, 5))
edges = cv2.Canny(image=blurred_img, threshold1=lower, threshold2=upper+50)
plt.imshow(edges)
plt.title("Edges with formula")
plt.show()
```
