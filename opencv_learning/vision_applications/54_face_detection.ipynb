```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def display(image, cmap='gray', title=None):
    plt.figure(figsize=(10, 6))
    plt.imshow(image, cmap=cmap)
    if title:
        plt.title(title)
    plt.show()
```

```python
nadia = cv2.imread("../../resources/DATA/Nadia_Murad.jpg", 0)
denis = cv2.imread("../../resources/DATA/Denis_Mukwege.jpg", 0)
solvay = cv2.imread("../../resources/DATA/solvay_conference.jpg", 0)
```

```python
display(solvay, title='Solvay Conference')
```

```python
face_classifier = cv2.CascadeClassifier("../../resources/DATA/haarcascades/haarcascade_frontalface_default.xml")
```

```python
def detect_image(img):
    face_img = img.copy()
    face_rects = face_classifier.detectMultiScale(face_img)

    for (x, y, w, h) in face_rects:
        cv2.rectangle(face_img, (x, y), (x+w, y+h), (255, 255, 255), 7)

    return face_img
```

```python
denis_result = detect_image(denis)
display(image=denis_result, title="Denis Face Detection")
```

```python
nadia_result = detect_image(nadia)
display(image=nadia_result, title="Nadia Face Detection")
```

```python
solvay_results = detect_image(solvay)
display(solvay_results, title="Solvay Conference Face Detection")
```

```python
def adjusted_detect_image(img):
    face_img = img.copy()
    face_rects = face_classifier.detectMultiScale(face_img, scaleFactor=1.2, minNeighbors=5)

    for (x, y, w, h) in face_rects:
        cv2.rectangle(face_img, (x, y), (x+w, y+h), (255, 255, 255), 7)

    return face_img
```

```python
solvay_results = adjusted_detect_image(solvay)
display(solvay_results, title="Solvay Conference Adjusted Face Detection")
```

```python
eye_cascade = cv2.CascadeClassifier("../../resources/DATA/haarcascades/haarcascade_eye.xml")
eye_cascade
```

```python
def detect_eyes(img):
    face_img = img.copy()
    eye_rects = eye_cascade.detectMultiScale(face_img, scaleFactor=1.2, minNeighbors=5)

    for (x, y, w, h) in eye_rects:
        cv2.rectangle(face_img, (x, y), (x+w, y+h), (255, 255, 255), 7)

    return face_img
```

```python
nadia_result = detect_eyes(nadia)
display(nadia_result, title="Eye Detection Nadia")
```

```python
denis_result = detect_eyes(denis)
display(denis_result, title="Eye Detection Denis")
```
