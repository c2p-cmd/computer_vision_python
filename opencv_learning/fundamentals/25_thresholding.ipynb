```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

doggy_img = cv2.imread('../../resources/DATA/00-puppy.jpg')
doggy_img = cv2.cvtColor(doggy_img, cv2.COLOR_BGR2RGB)

print(doggy_img.shape)
plt.imshow(doggy_img)
plt.axis('off')
plt.title("Puppy")
plt.show()
```

```python
import seaborn as sns

r = doggy_img[:, :, 0].flatten()
g = doggy_img[:, :, 1].flatten()
b = doggy_img[:, :, 2].flatten()

plt.figure(figsize=(10, 6))
sns.histplot(r, label='R', color='red', alpha=0.5, kde=True, bins=50)
sns.histplot(g, label='G', color='green', alpha=0.5, kde=True, bins=50)
sns.histplot(b, label='B', color='blue', alpha=0.5, kde=True, bins=50)
plt.grid()
plt.legend()
plt.title("Image Histogram")
plt.show()
```

```python
doggy_img_gray = cv2.cvtColor(doggy_img, cv2.COLOR_RGB2GRAY)
doggy_img_gray_flat = doggy_img_gray.flatten()

plt.figure(figsize=(10, 6))
sns.histplot(doggy_img_gray_flat, alpha=0.5, kde=True, bins=50)
plt.grid()
plt.title("Image GrayScale Histogram")
plt.show()
```

```python
crossword = cv2.imread('../../resources/DATA/crossword.jpg', 0)
plt.title("Crossword")
plt.imshow(crossword, cmap='gray')
plt.show()
```

```python
def show_pic(picture):
    fig = plt.figure(figsize=(15, 15))
    ax = fig.add_subplot(111)
    ax.imshow(picture, cmap='gray')

show_pic(picture=crossword)
```

```python
ret, thres2 = cv2.threshold(crossword, 180, 255, cv2.THRESH_BINARY)
show_pic(thres2)
```

```python
th2 = cv2.adaptiveThreshold(crossword, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 7, 8)
show_pic(th2)
```

```python
blended = cv2.addWeighted(src1=thres2, alpha=0.6, src2=th2, beta=0.4, gamma=0)
show_pic(blended)
```
