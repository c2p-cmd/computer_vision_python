```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blank_img = np.zeros(shape=(512, 512, 3), dtype=np.int16)
blank_img
```

## Base Image

```python
plt.title("Base Blank Image")
plt.imshow(blank_img)
plt.show()
```

### Making Rectangles

```python
blank_img = cv2.rectangle(
    img=blank_img,
    pt1=(350, 10),
    pt2=(400, 150),
    color=(0, 255, 128),
    thickness=5
)
plt.imshow(blank_img)
plt.title('Rectangle')
```

```python
blank_img = cv2.rectangle(
    blank_img,
    pt1=(200, 200),
    pt2=(300, 300),
    color=(0,0,255),
    thickness=3
)
plt.imshow(blank_img)
plt.title("Center")
```

## Drawing Circles (outlined and filled)

```python
blank_img = cv2.circle(
    img=blank_img,
    center=(100,100),
    radius=50,
    color=(255,100,50),
    thickness=5
)
plt.imshow(blank_img)
plt.title("Outlined Circle")
plt.show()
```

```python
blank_img = cv2.circle(
    img=blank_img,
    center=(250, 400),
    radius=50,
    color=(255,100,50),
    thickness=-1
)
plt.imshow(blank_img)
plt.title("Filled Circle")
plt.show()
```

## Drawing Lines

```python
blank_img = cv2.line(
    img=blank_img,
    pt1=(0,0),
    pt2=(512,512),
    color=(200, 55, 55),
    thickness=7
)
blank_img = cv2.line(
    img=blank_img,
    pt1=(0,512),
    pt2=(512,0),
    color=(200, 55, 55),
    thickness=7
)
plt.imshow(blank_img)
plt.title("X-Mark with 2 lines")
plt.show()
```

## Drawing Polygons

```python
blank_img = np.zeros_like(blank_img)

plt.title("Blank")
plt.imshow(blank_img)
plt.show()
```

```python
vertices = np.array([
    [100, 300],
    [250, 250],
    [300, 400],
], dtype=np.int32)
vertices
```

```python
points = np.expand_dims(vertices, 1)
points
```

```python
from IPython.display import display, Markdown

Markdown(f"#### Vertices Shape: `{vertices.shape}`, Points Shape: `{points.shape}`")
```

```python
blank_img = cv2.polylines(blank_img, [points], isClosed=True, color=(200, 55, 55), thickness=7)

plt.title("Polygon on Image")
plt.imshow(blank_img)
plt.show()
```

```python
blank_img = np.zeros_like(blank_img)
blank_img = cv2.fillPoly(blank_img, [points], color=(100, 55, 255))

plt.title("Filled Polygon on Image")
plt.imshow(blank_img)
plt.show()
```

```python
blank_img = cv2.putText(
    blank_img,
    text='Triangle',
    org=(250, 180),
    fontFace=cv2.FONT_HERSHEY_SCRIPT_SIMPLEX,
    fontScale=2,
    color=(255, 255, 255),
    thickness=3,
    lineType=cv2.LINE_AA
)

plt.title("Adding Text")
plt.imshow(blank_img)
plt.show()
```
